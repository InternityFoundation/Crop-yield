```python
df= we[['Date','TempAvgF', 'DewPointAvgF', 'HumidityAvgPercent', 'SeaLevelPressureAvgInches', 'Events']]
```

```python
df.head()
```

```python
df=df.drop(["Date"],axis=1)
```

```python
df.head()
```

```python
df['Events'] = df['Events'].replace(' ', 'Clear')
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder

# encoding
preprocessed_data = df
encoder = LabelEncoder()
encoder.fit(df['Events'])
preprocessed_data['Events'] = encoder.transform(df['Events'])

# printing
classes_occurrences = preprocessed_data['Events'].value_counts().to_frame()
classes_occurrences['Class'] = encoder.inverse_transform(classes_occurrences.index)
classes_occurrences = classes_occurrences.sort_index(axis=0)
print(classes_occurrences)
```

```python
df['Events'] = df['Events'].replace([1], 0)
df['Events'] = df['Events'].replace([2,5,6], 1)
df['Events'] = df['Events'].replace([3,4,7,8], 2)

# printing classes and number of occurrences
classes_occurrences = df['Events'].value_counts().to_frame()
classes_occurrences['Class'] = ['Clear', 'Rain', 'Thunderstorm']
classes_occurrences = classes_occurrences.sort_index(axis=0)
print(classes_occurrences)
```

```python
indexNames = df[ df['DewPointAvgF'] == '-' ].index
 
# Delete these row indexes from dataFrame
df.drop(indexNames , inplace=True)
```

```python
from sklearn.model_selection import train_test_split 
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, 
                                                    random_state=2018)

clf = DecisionTreeClassifier(random_state=2018)

#clf = DecisionTreeClassifier(max_depth=3,random_state=2018,
#                            min_samples_split=20,min_samples_leaf=5)
clf.fit(X_train,y_train)

y_pred = clf.predict(X_test)
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
print(accuracy_score(y_test,y_pred))
```
